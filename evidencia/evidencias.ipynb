```python
usuarios = ["Ana", "Luis"]
roles = ["Lector", "Administrador"]

for rol_ana in roles:
    for rol_luis in roles:

        administradores = 0

        if rol_ana == "Administrador":
            administradores += 1

        if rol_luis == "Administrador":
            administradores += 1

        if administradores <= 1:

            print("Solución válida:")
            print("Ana ->", rol_ana)
            print("Luis ->", rol_luis)
            print()
```
